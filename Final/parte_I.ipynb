# DataConnect: Análise de Indicadores Educacionais e Tecnológicos

Neste projeto, exploramos três DataFrames principais que fornecem informações valiosas sobre indicadores educacionais e tecnológicos em diferentes entidades ao redor do mundo. Cada DataFrame possui colunas específicas que foram exploradas durante a análise.

## DataFrame `df_conexao_internet`

Nome das colunas:
* Entity: Entidade - Informações sobre o código do país e quando o código tem o valor 'Região', denota divisão agrupando vários países.
* Code: Código - Informações sobre o código do país e quando o código tem o valor 'Região', denota divisão agrupando vários países.
* Year: Ano
* Cellular Subscription: Assinatura de Celular - Assinaturas de telefones celulares por 100 pessoas. Esse número pode ultrapassar 100 quando a pessoa média tem mais de uma assinatura de serviço móvel.
* Internet Users(%): Usuários de Internet (%) - A parcela da população que acessa a internet em todos os países do mundo.
* No. of Internet Users: Número de Usuários de Internet - Número de pessoas usando a Internet em cada país.
* Broadband Subscription: Assinatura de Banda Larga - O número de assinaturas de banda larga fixa por 100 pessoas. Isso se refere a assinaturas fixas para acesso de alta velocidade à Internet pública (uma conexão TCP/IP), em velocidades downstream iguais ou superiores a 256 kbit/s.

Este DataFrame apresenta indicadores relacionados à conectividade e uso da internet, incluindo assinatura de celular, usuários de internet (%), número total de usuários de internet e assinatura de banda larga. Algumas ações realizadas incluíram:

- Verificação de valores nulos.
- Exibição dos primeiros e últimos valores.
- Descrição estatística das colunas.

[_Fonte_](https://www.kaggle.com/datasets/ashishraut64/internet-users/data)

```python
# Inicializa o tradutor
tradutor = Translator()

# Função para traduzir uma única localidade
def traduzir_localidade(localidade):
    try:
        return tradutor.translate(localidade, dest='pt').text
    except Exception as e:
        print(f"Erro ao traduzir {localidade}: {e}")
        return localidade

# Aplica a tradução à coluna 'Localidade' do DataFrame
df_educacao_na_rede['Localidade'] = df_educacao_na_rede['Localidade'].apply(traduzir_localidade)

# Exibe o DataFrame com a coluna 'Localidade' traduzida
print(df_educacao_na_rede)
```

```python
# Definindo o estilo de plotagem como 'dark_background'
plt.style.use('dark_background')

# Lista de países da América do Sul
paises_america_sul = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela']

# Filtrar o DataFrame para incluir apenas os países da América do Sul
df_america_sul = df_educacao_na_rede[df_educacao_na_rede['Localidade'].isin(paises_america_sul)]

# Filtrar dados para os anos desejados (2015 a 2021)
df_filtrado_america_sul = df_america_sul[(df_america_sul['Ano'] >= 2015) & (df_america_sul['Ano'] <= 2021)]

# Agrupar por localidade e assinatura de banda larga
casas_com_acesso_america_sul = df_filtrado_america_sul.groupby(['Localidade', 'Ano', 'Assinatura de Banda Larga'])['Código'].count().unstack()

# Somar casas com e sem acesso à banda larga
casas_com_acesso_america_sul['Total'] = casas_com_acesso_america_sul.sum(axis=1)

# Ordenar as localidades pela média do total
localidades_mais_dados_america_sul = casas_com_acesso_america_sul.groupby('Localidade')['Total'].mean().sort_values(ascending=False).head(10).index

# Filtrar para as 10 localidades com mais dados
casas_com_acesso_top10_america_sul = casas_com_acesso_america_sul.loc[localidades_mais_dados_america_sul]

# Plotar gráfico de barras empilhadas
ax = casas_com_acesso_top10_america_sul.drop('Total', axis=1).plot(kind='bar', stacked=True, xlabel='Localidade', ylabel='Número de Casas', title='Crescimento de Casas com Acesso à Internet (Top 10 Localidades - América do Sul)')

# Ajustar a posição da legenda
leg = ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Ano')

plt.show()
```
